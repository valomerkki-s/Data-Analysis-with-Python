e) Create a categorical version of the revenue column with 4 groups: (i) values between 0 and 3500, (ii)
values between 3500 and 6380, (iii) values between 6380 and 14620, and (iv) values between 14620
and 1000000. Label the groups [’Low’, ’Medium’, ’High’, ’Very High’]. In which category do you find
the most companies?
• Which sector has the highest average revenue?

```python
revenue_bins = [0, 3500, 6380, 14620, 1000000]
labels = ['Low', 'Medium', 'High', 'Very High']
fortune_df['rev_categ'] = pd.cut(fortune_df.revenue, revenue_bins, labels = labels)
```

```python
fortune_df.rev_categ.value_counts()

#there is no revenue category tat stands out as the one to which most company belong to
# in fact all the 4 bins contain the same amount of companes.
# But since also the given endpoints for revenue ranges are not of equal size, this output is expectable
# it is almost as if the 1000 list has ben divided in 4 equal parts,
#and the revenue of each part's last element has been used for the given endpoints...
```

```python
sectors = fortune_df.sector.unique()
sectors

#this will display all the sectors
```

```python
sec_revs = pd.Series(dtype = float, index = sectors)

#this creates a series with as index the above collected sectors
```

```python
for i in sectors:
    sec_revs[i] = fortune_df[fortune_df.sector == i].revenue.mean()
    
sec_revs.sort_values(ascending = False)

#and this appends in the Series the mean of each sector revenue from the original dataframe.
#the sector with the highest revenue is Telecommunications.
#sec_revs.sort_values(ascending = False)[:1] can also be used
#to display (and access) only the first item of the list, but i prefered seeing them all for sake of clarity
```
